# Detect Kernel Language
Use IPython or Jupyter APIs to detect and display the current kernel language.

```python
try:
    from IPython import get_ipython
    kernel = get_ipython()
    print('Kernel language:', kernel.language_info['name'])
except Exception as e:
    print('Could not detect kernel language:', e)
```

# Display Language Information
Show additional information about the language, such as sys.executable and platform details.

```python
import sys, platform
print('sys.executable:', sys.executable)
print('platform:', platform.platform())
```
